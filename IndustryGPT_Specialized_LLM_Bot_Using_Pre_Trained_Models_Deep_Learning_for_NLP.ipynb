```python
import requests
from bs4 import BeautifulSoup

def scrape_techcrunch_article(url):
    """
    Scrapes the main content from a TechCrunch article.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx or 5xx)
        soup = BeautifulSoup(response.text, 'lxml')

        # TechCrunch article content is typically within a div with class 'article-content'
        # or similar. We need to inspect the current HTML structure.
        # Let's try to find the main article body.
        article_body = soup.find('div', class_='article-content')
        if not article_body:
            article_body = soup.find('div', class_='entry-content') # Another common class
        if not article_body:
            article_body = soup.find('section', class_='article-main__content')
        if not article_body:
            # If article body not found, try to get some content from the main page as a fallback
            print(f"Could not find article body specific tag for {url}, attempting to extract general text.")
            paragraphs = soup.find_all('p')
            article_text = '\n'.join([p.get_text() for p in paragraphs])
            return article_text.strip() if article_text else None

        paragraphs = article_body.find_all('p')
        article_text = '\n'.join([p.get_text() for p in paragraphs])
        return article_text.strip()

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None

def scrape_aws_documentation(url):
    """
    Placeholder function to scrape content from AWS documentation.
    Actual implementation would involve inspecting AWS docs structure.
    """
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'lxml')
        # AWS documentation often has content within main divs or articles
        main_content = soup.find('div', {'id': 'main-col-body'}) # Common for AWS docs
        if not main_content:
            main_content = soup.find('article')
        if not main_content:
            print(f"Could not find main content for AWS URL: {url}")
            return None

        paragraphs = main_content.find_all('p')
        aws_text = '\n'.join([p.get_text() for p in paragraphs])
        return aws_text.strip()

    except requests.exceptions.RequestException as e:
        print(f"Error fetching AWS documentation {url}: {e}")
        return None

# Example usage for TechCrunch - trying a generic homepage URL to ensure content retrieval
techcrunch_url = 'https://techcrunch.com/'
print(f"Attempting to scrape: {techcrunch_url}")
techcrunch_content = scrape_techcrunch_article(techcrunch_url)

if techcrunch_content:
    print("\n--- TechCrunch Homepage Content (first 500 chars) ---")
    print(techcrunch_content[:500])
    print("\n--- End of TechCrunch sample ---")
else:
    print("Failed to scrape TechCrunch content.")

# Now proceed to AWS Documentation scraping
aws_doc_url = 'https://docs.aws.amazon.com/lambda/latest/dg/welcome.html' # A generic AWS Lambda doc page
print(f"\nAttempting to scrape AWS Documentation: {aws_doc_url}")
aws_content = scrape_aws_documentation(aws_doc_url)

if aws_content:
    print("\n--- AWS Documentation Content (first 500 chars) ---")
    print(aws_content[:500])
    print("\n--- End of AWS sample ---")
else:
    print("Failed to scrape AWS documentation content.")

# Placeholder for tech blog scraping
print("\nFurther steps will involve scraping tech blogs.")
# medium_url = 'https://medium.com/mlearning-ai/a-comprehensive-guide-to-llm-fine-tuning-for-specific-tasks-9f7e7a1b7e6e'
# medium_content = scrape_medium_blog(medium_url)
```

## Model Training and Fine-tuning


Fine-tune the selected pre-trained model (`DistilBERT-base-uncased`) using the collected 'Information and Technology' data. Utilize Google Colab with T4 GPUs, limiting the training to a maximum of 25 epochs.



The first step for model training and fine-tuning is to install the required libraries, `transformers` and `datasets`, which are essential for working with Hugging Face models and datasets.


Now that the necessary libraries are installed, the next step is to load the previously collected textual data from the files, prepare it for fine-tuning, instantiate the `DistilBERT` model and tokenizer, define training arguments, and initiate the fine-tuning process using the Hugging Face `Trainer` API.



```python
import os
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import torch

# 1. Load the previously collected text data
def load_text_from_file(filename):
    """Loads content from a text file, handles file not found."""
    if not os.path.exists(filename):
        print(f"Warning: File not found: {filename}. Returning empty string.")
        return ""
    try:
        with open(filename, 'r', encoding='utf-8') as f:
            return f.read()
    except Exception as e:
        print(f"Error reading file {filename}: {e}. Returning empty string.")
        return ""

techcrunch_content = load_text_from_file('techcrunch_homepage_content.txt')
aws_content = load_text_from_file('aws_lambda_doc_content.txt')

# 2. Combine these texts into a single string, separating chunks by double newline
combined_raw_text = techcrunch_content + "\n\n" + aws_content

# --- Prepare data for fine-tuning ---
# Instantiate the tokenizer
model_name = "distilbert-base-uncased"
tokenizer = DistilBertTokenizerFast.from_pretrained(model_name)

# Split the combined text into smaller chunks (e.g., by paragraph) to create multiple training examples.
# This helps in handling long documents and creating more diverse training instances.
# Filter out very short chunks that might not contain meaningful information.
raw_texts_chunks = [chunk.strip() for chunk in combined_raw_text.split('\n\n') if len(chunk.strip()) > 50]

# 4. Tokenize the text data chunks
# Max sequence length for DistilBERT is 512. Using a smaller length for demonstration
# to fit more examples in a batch and speed up training.
max_seq_length = 128 # Can be increased up to 512 if resources allow and for better context

tokenized_data = tokenizer(
    raw_texts_chunks,
    truncation=True,
    padding='max_length',
    max_length=max_seq_length,
    return_tensors='pt' # Return PyTorch tensors
)

# 5. Prepare a Dataset object
# For DistilBertForSequenceClassification, we need input_ids, attention_mask, and labels.
# Since we don't have specific labels for a classification task based on raw text,
# we will create dummy labels (e.g., all zeros) for a single-class classification setup.
# This primarily trains the model to adapt its embeddings to the domain's language patterns.
# In a real-world scenario for an LLM Bot, true fine-tuning might involve Masked Language Modeling (MLM)
# for domain adaptation, or supervised fine-tuning on Question Answering (QA) datasets.
num_labels_for_dummy_task = 1
# Fix: Convert dummy_labels to float if num_labels_for_dummy_task is 1 (regression-like task)
# Otherwise, keep it as long for multi-class classification.
if num_labels_for_dummy_task == 1:
    dummy_labels = torch.zeros(len(raw_texts_chunks), dtype=torch.float)
else:
    dummy_labels = torch.zeros(len(raw_texts_chunks), dtype=torch.long)

# Convert to Hugging Face Dataset format
hf_dataset = Dataset.from_dict({
    'input_ids': tokenized_data['input_ids'],
    'attention_mask': tokenized_data['attention_mask'],
    'labels': dummy_labels, # Add dummy labels
})

# 3. Instantiate DistilBertForSequenceClassification
# num_labels should match the number of unique labels in our dataset (which is 1 here, for label 0)
model = DistilBertForSequenceClassification.from_pretrained(model_name, num_labels=num_labels_for_dummy_task)

# Check for GPU availability and move model to GPU if present
if torch.cuda.is_available():
    model.to('cuda')
    print(f"Model moved to {model.device} (GPU).")
else:
    print("CUDA not available. Model will train on CPU. For faster training, ensure a GPU runtime is enabled.")

# 6. Define training arguments
output_dir = './distilbert_it_finetuned'
training_args = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=1,  # Setting to 1 epoch for demonstration. Subtask allows up to 25 epochs.
    per_device_train_batch_size=16, # Adjust based on GPU memory. Common values: 8, 16, 32.
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=50,
    save_strategy="no", # Not saving checkpoints for this quick demonstration
    report_to="none", # Disable reporting to external services
    fp16=torch.cuda.is_available(), # Use mixed precision training if GPU is available
    dataloader_num_workers=2 # Increase if data loading is a bottleneck on GPU
)

# 7. Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=hf_dataset,
)

# 8. Start the fine-tuning process
print(f"Starting fine-tuning of {model_name} on {len(hf_dataset)} examples for {training_args.num_train_epochs} epoch(s)...")
trainer.train()
print("Fine-tuning process completed.")
```


The previous code successfully completed the fine-tuning process for one epoch. To complete the fine-tuning subtask and make the trained model available for future steps, it is essential to save the fine-tuned model and its tokenizer.



```python
print("Saving the fine-tuned model and tokenizer...")
# Save the fine-tuned model and tokenizer
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)

print(f"Fine-tuned model and tokenizer saved to: {output_dir}")
print("Model fine-tuning and saving completed. Ready for evaluation or further development.")
```

## LLM Bot Development


Developing the LLM Bot to interact with users, providing answers and engaging in meaningful conversations specific to the 'Information and Technology' industry. The bot should demonstrate the ability to understand and process industry-related queries effectively.



The first step in developing the LLM Bot is to load the fine-tuned `DistilBERT` model and its tokenizer, which were saved in the previous step, to prepare for embedding generation and retrieval.



```python
import os
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification
import torch
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Define the output directory where the fine-tuned model and tokenizer were saved
output_dir = './distilbert_it_finetuned'

print(f"Loading fine-tuned model and tokenizer from: {output_dir}")

# Load the tokenizer
tokenizer = DistilBertTokenizerFast.from_pretrained(output_dir)

# Load the model
# We load it as DistilBertForSequenceClassification because that's how it was saved.
# We will extract the base model's embeddings for retrieval purposes.
model = DistilBertForSequenceClassification.from_pretrained(output_dir)

# Move model to GPU if available
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)
model.eval() # Set model to evaluation mode

print("Fine-tuned model and tokenizer loaded successfully.")
print(f"Model is running on: {device}")
```


The next step is to generate embeddings for each chunk in the `raw_texts_chunks` knowledge base using the loaded `DistilBERT` model. This will create a numerical representation for each text piece, which is essential for similarity search.



```python
def get_embedding(text, tokenizer, model, device, max_length=128):
    """
    Generates an embedding for a given text using the DistilBERT model.
    The model was fine-tuned for sequence classification, so we will extract
    the embedding from its base model's last hidden state, typically the [CLS] token's representation.
    """
    # Ensure the text is a list for batch processing, even if single
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding='max_length', max_length=max_length)
    inputs = {key: val.to(device) for key, val in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs, output_hidden_states=True)

    # Extract the [CLS] token's embedding from the last layer
    # The 'base_model' refers to DistilBERTModel which outputs hidden states
    # The last hidden state is at outputs.hidden_states[-1]
    # The [CLS] token is the first token (index 0)
    cls_embedding = outputs.hidden_states[-1][:, 0, :]
    return cls_embedding.cpu().numpy()

# Generate embeddings for the knowledge base chunks
knowledge_base_embeddings = []
for i, chunk in enumerate(raw_texts_chunks):
    if i % 10 == 0: # Print progress every 10 chunks
        print(f"Generating embedding for chunk {i+1}/{len(raw_texts_chunks)}")
    embedding = get_embedding(chunk, tokenizer, model, device, max_length=max_seq_length)
    knowledge_base_embeddings.append(embedding.flatten()) # Flatten the 1D tensor for easier storage

knowledge_base_embeddings = np.array(knowledge_base_embeddings)

print(f"Generated embeddings for {len(raw_texts_chunks)} knowledge base chunks.")
print(f"Shape of knowledge_base_embeddings: {knowledge_base_embeddings.shape}")
```


With the knowledge base embeddings generated, the next step is to implement a function to process user queries by generating their embeddings and finding the most similar chunks from the knowledge base using cosine similarity, as outlined in instruction #4 of the subtask.



```python
def retrieve_relevant_chunks(query, tokenizer, model, device, knowledge_base_embeddings, raw_texts_chunks, top_n=2, max_length=max_seq_length):
    """
    Generates an embedding for the query and retrieves the top_n most similar chunks
    from the knowledge base using cosine similarity.
    """
    query_embedding = get_embedding(query, tokenizer, model, device, max_length=max_length).flatten()

    # Calculate cosine similarity between query embedding and all knowledge base embeddings
    similarities = cosine_similarity(query_embedding.reshape(1, -1), knowledge_base_embeddings)

    # Get indices of top_n most similar chunks
    top_n_indices = similarities.argsort()[0][::-1][:top_n]

    # Retrieve the actual text chunks
    relevant_chunks = [raw_texts_chunks[i] for i in top_n_indices]

    print(f"\nQuery: {query}")
    print(f"Top {top_n} most relevant chunks found:")
    for i, chunk in enumerate(relevant_chunks):
        print(f"""---
Chunk {i+1}:
{chunk[:500]}...""") # Print first 500 chars of the chunk

    return relevant_chunks

print("Function 'retrieve_relevant_chunks' defined. Ready for bot development.")
```


With the `retrieve_relevant_chunks` function defined, the next step is to implement a simple conversational loop as outlined in instruction #5, allowing users to input queries and demonstrating the bot's ability to retrieve relevant information based on instruction #6 of the subtask.



```python
print("\n--- LLM Bot Conversation (Type 'quit' to exit) ---\n")

while True:
    user_query = input("You: ")
    if user_query.lower() == 'quit':
        print("Bot: Goodbye!")
        break

    if not user_query.strip():
        print("Bot: Please enter a query.")
        continue

    # Retrieve relevant chunks based on the user's query
    relevant_chunks = retrieve_relevant_chunks(user_query, tokenizer, model, device, knowledge_base_embeddings, raw_texts_chunks, top_n=2, max_length=max_seq_length)

    print("\nBot's Response (based on retrieved knowledge):")
    if relevant_chunks:
        for i, chunk in enumerate(relevant_chunks):
            print(f"--- Source Chunk {i+1} ---")
            print(chunk)
            print("----------------------")
    else:
        print("Bot: I couldn't find any relevant information for your query in my knowledge base.")
    print("\n")

print("Bot development and testing complete for this iteration.")
```
